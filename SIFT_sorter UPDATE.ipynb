```python
import numpy as np

file_path = 'C:/Users/kfc/Desktop/SIFT/SIFT_GLUT13.txt'

results = []

with open(file_path, 'r') as file:
    lines = file.readlines()

for line in lines[9:]:
    row = line.strip().split('\t')

    if len(row) >= 11:  
        label = row[0]  
        values = row[1:]
        try:
            values = [float(value) for value in values]  
            average = np.mean(values)
            row_result = [label, average]
            results.append(row_result)
        except ValueError:
            pass  

for row_result in results:
    print(row_result[0], row_result[1])
    
output_file_path = file_path.rsplit('.', 1)[0] + '_ave.txt'
with open(output_file_path, 'w') as output_file:
    for row_result in results:
        output_file.write(f"{row_result[0]} {row_result[1]}\n")
```
