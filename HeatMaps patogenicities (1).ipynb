```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the .txt file into a DataFrame
data = pd.read_csv("D:/GLUTtransporters/patogenicities.txt", delimiter="\t").set_index("ID")
data.head()

input_file_path = "D:/GLUTtransporters/patogenicities.txt"
output_directory = os.path.dirname(os.path.abspath(input_file_path))

# Create a heatmap for the second column with the 0 to 1 scale and blue to red color map
plt.figure(figsize=(8, 6))
sns.heatmap(data.iloc[:, [0]], cmap="RdYlBu_r", vmin=0, vmax=1, cbar=True, annot=True, fmt=".2f")
plt.title("AlphaMissense")
plt.xlabel("Average Pathogenicty")
plt.ylabel("Protein")
output_file_path = os.path.join(output_directory, "AlphaMissense_heatmap.svg")
plt.savefig(output_file_path)
plt.show()

# Create a heatmap for the third column with the -1 to 0 scale and blue to red color map
plt.figure(figsize=(8, 6))
sns.heatmap(data.iloc[:, [1]], cmap="RdYlBu", vmin=0.1, vmax=0.2, cbar=True, annot=True, fmt=".2f", yticklabels = True)
plt.title("SIFT")
plt.xlabel("Value")
plt.ylabel("Protein")
output_file_path = os.path.join(output_directory, "SIFT_heatmap.svg")
plt.savefig(output_file_path)
plt.show()

# Create a heatmap for the fourth column with the 0 to 1 scale and blue to red color map
plt.figure(figsize=(8, 6))
sns.heatmap(data.iloc[:, [2]], cmap="RdYlBu_r", vmin=0, vmax=1, cbar=True, annot=True, fmt=".2f")
plt.title("Poly Phen 2")
plt.xlabel("Value")
plt.ylabel("Protein")
output_file_path = os.path.join(output_directory, "Poly_Phen_2_heatmap.svg")
plt.savefig(output_file_path)
plt.show()
# Save the heatmap as a .png file in the same location as the input file
#input_file_path = "D:/GLUTtransporters/patogenicities.txt"
#output_file_path = input_file_path.replace(".txt", "_heatmap.svg")
#plt.savefig(output_file_path)
#plt.show()
```
