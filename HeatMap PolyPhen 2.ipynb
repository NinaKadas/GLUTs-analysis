```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("D:/Rhapsody/Rhapsody/GLUT13_ave_poly.csv", delimiter=";")

# Create an empty dictionary to store the sums and counts for each letter
letter_data = {}

# Iterate through the rows of the DataFrame
for index, row in df.iterrows():
    letter = row['Label.1']
    number = row['PolyPhen-2']

    if letter in letter_data:
        letter_data[letter]['sum'] += number
        letter_data[letter]['count'] += 1
    else:
        letter_data[letter] = {'sum': number, 'count': 1}

# Calculate the averages and store the results in a list
letter_averages = []
for letter, data in letter_data.items():
    average = data['sum'] / data['count']
    letter_averages.append({'Letter': letter, 'Average': average})

# Create a DataFrame from the averages
average_df = pd.DataFrame(letter_averages)

# Create a pivot table to make it suitable for a heatmap
heatmap_data = average_df.pivot_table(index='Letter', values='Average')

# Increase the figure size to fit the entire heatmap
plt.figure(figsize=(3, 6))  # Adjust the width (10) as needed

# Create the heatmap
sns.heatmap(heatmap_data, cmap="RdYlBu_r", annot=True, fmt=".2f", cbar_kws={'label': 'Average'})
plt.title('Average pathogenicities of amino acids')

# Save the heatmap as a .png file in the same location as the input file
input_file_path = "D:/Rhapsody/Rhapsody/GLUT13_ave_poly.csv"
output_file_path = input_file_path.replace(".csv", "_heatmap.png")
plt.savefig(output_file_path)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Initialize a dictionary to store the sum and count of averages for each amino acid
amino_acid_avg_dict = {}

# Directory containing your CSV files
directory = r'C:\Users\kfc\Desktop\GLUTPolyPhen-2'

input_file_path = r"C:\Users\kfc\Desktop\GLUTPolyPhen-2"
output_directory = os.path.dirname(os.path.abspath(input_file_path))

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):  # Process only CSV files
        file_path = os.path.join(directory, filename)
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path, delimiter=";")
        # Create an empty dictionary to store the sums and counts for each letter
        letter_data = {}
        # Iterate through the rows of the DataFrame
        for index, row in df.iterrows():
            letter = row['Label.1']
            number = row['PolyPhen-2']
            if letter in letter_data:
                letter_data[letter]['sum'] += number
                letter_data[letter]['count'] += 1
            else:
                letter_data[letter] = {'sum': number, 'count': 1}
        # Calculate the averages and store the results in a dictionary
        for letter, data in letter_data.items():
            average = data['sum'] / data['count']
            if letter in amino_acid_avg_dict:
                amino_acid_avg_dict[letter].append(average)
            else:
                amino_acid_avg_dict[letter] = [average]

# Calculate the average of averages for each amino acid
combined_avg_dict = {}
for amino_acid, avg_list in amino_acid_avg_dict.items():
    combined_avg_dict[amino_acid] = round(sum(avg_list) / len(avg_list), 2)

# Convert the result into a DataFrame
combined_avg_df = pd.DataFrame(list(combined_avg_dict.items()), columns=['Letter', 'Average'])

# Create a heatmap using Seaborn with annotations
plt.figure(figsize=(8, 6))
ax = sns.heatmap(combined_avg_df.pivot_table(index='Letter', values='Average'), cmap='RdYlBu_r', annot=True, fmt=".2f", vmin=0, vmax=1)
#plt.xlabel('Combined Average')
plt.ylabel('Amino Acids')
plt.title('Poly Phen 2')
output_file_path = os.path.join(output_directory, "Poly Phen 2_AMK_heatmap.svg")
plt.savefig(output_file_path)
# Display the heatmap
plt.show()
```
