```python
import pandas as pd

# Read the .csv file
df = pd.read_csv('D:/ClinVar/GLUT1.csv', delimiter=';')

# Extract the numbers from the 'Protein change' column
df['Number'] = df['Protein change'].str.extract('(\d+)').astype(int)

# Initialize an empty dictionary to store the results
results = {}

# Open the .pdb file
with open('D:/GLUTnewpymissense/P11166-edit.pdb', 'r') as f:
    for line in f:
        # Only consider lines that start with 'ATOM'
        if line.startswith('ATOM'):
            # Split the line into columns
            columns = line.split()
            # Check if the sixth column is a number
            if columns[5].isdigit():
                # Convert the sixth column to an integer
                number = int(columns[5])
                # Check if the number is in the 'Number' column of the dataframe
                if number in df['Number'].values:
                    # If the number is found, store the eleventh column in the results dictionary
                    # Use the first occurrence of the number
                    if number not in results:
                        results[number] = columns[10]

# Print the results
for protein_change, number in zip(df['Protein change'], df['Number']):
    if number in results:
        print(f'{protein_change}: {results[number]}')
```
