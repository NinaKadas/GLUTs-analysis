```python
import pandas as pd

# Load the first CSV file
df1 = pd.read_csv('C:/Users/kfc/Desktop/ClinVar/GLUT9.csv', delimiter=';')

# Load the second CSV file
df2 = pd.read_csv('D:/SIFT/GLUT9.csv', delimiter=';')

# Initialize an empty list to store the results
results = []

# Iterate over each row in the first DataFrame
for index, row in df1.iterrows():
    # Check if the 'Protein Change' value is a string
    if isinstance(row['Protein change'], str):
        # Split the 'Protein Change' value into prefix, number, and suffix
        protein_change = row['Protein change']
        prefix = protein_change[0]
        number = protein_change[1:-1]
        suffix = protein_change[-1]

        # Find the corresponding row in the second DataFrame
        df2_row = df2[df2['pos'] == number + prefix]

        # If the row exists, get the value from the column named after the suffix
        if not df2_row.empty:
            value = df2_row[suffix].values[0]
            results.append((protein_change, value))

# Print the results
for protein_change, value in results:
    print(f'{protein_change}: {value}')
```
